Systematic trading does not begin with models. It begins with **data discipline**.

Before signals, before backtests, before performance attribution, every systematic strategy rests on assumptions encoded in its data.

## 1. Core Principles of Market Data

All market data falls into three conceptual layers:

1. Price and volume
2. State variables
3. Exogenous signals

## 2. Market Price Data (OHLCV)

Open, high, low, close, and volume data form the backbone of almost every strategy.

```
import yfinance as yf

data = yf.download('SPY', start='2000-01-01', end='2024-12-31')
data.head()
```
